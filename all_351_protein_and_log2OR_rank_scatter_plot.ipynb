# rep1/2 m6A log2OR corralation scatter plot

```python
import pandas as pd

pro_rep12_log2OR_data_file = "all_351_protein_and_rep12_log2OR_for_correlation_scatter_plot.txt"
pro_rep12_log2OR_df = pd.read_csv(pro_rep12_log2OR_data_file, sep="\t", header=0)
print(pro_rep12_log2OR_df.shape)
pro_rep12_log2OR_df.head()

```

```python
import numpy as np 
from scipy import stats
import matplotlib.pyplot as plt
rep1_m6A_log2OR = pro_rep12_log2OR_df.rep1_m6A_log2OR
rep2_m6A_log2OR = pro_rep12_log2OR_df.rep2_m6A_log2OR
pearsonr, pval = stats.pearsonr(rep1_m6A_log2OR, rep2_m6A_log2OR)

xy = np.vstack([rep1_m6A_log2OR, rep2_m6A_log2OR])
kde = stats.gaussian_kde(xy)(xy)

fig = plt.figure(figsize=(6, 6))
# ax1
ax1 = fig.add_subplot(1, 1, 1)
# ax1.axhline(y=0, color='grey', linestyle='--', lw=0.5, alpha=0.5)
# ax1.axvline(x=0, color='grey', linestyle='--', lw=0.5, alpha=0.5)
ax1.plot([-8, 8], [-8, 8], ls='--', lw=1, c="grey", alpha=0.4)
ax1.scatter(x=rep1_m6A_log2OR, y=rep2_m6A_log2OR, s=12, c=kde, cmap='coolwarm', edgecolors='none')
ax1.text(-7.5, 6.5, r'$ r = $'+str(format(pearsonr, '.6f')+"\n"+r'$ p = $'+str(format(pval, '.2e'))))
ax1.set_xlim(-8, 8)
ax1.set_ylim(-8, 8)
ax1.set_xlabel("rep1_m6A_log2OR")
ax1.set_ylabel("rep2_m6A_log2OR")

fig.savefig('all_351_protein_and_rep12_log2OR_correlation_scatter_plot.color_use_kde.pdf')
plt.show()

```
